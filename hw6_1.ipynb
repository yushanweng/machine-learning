```python
#def function 
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

def Supervised(): 
    for i in range(0,30):
        b_train_df, b_test_df = train_test_split(df_re_b, test_size = 0.2,random_state=0)
        m_train_df, m_test_df = train_test_split(df_re_m, test_size = 0.2,random_state=0)

        train_df = pd.concat([b_train_df, m_train_df], axis=0) #vertical concat
        test_df = pd.concat([b_test_df, m_test_df], axis=0)

        x_train_df_normalized = preprocessing.normalize(train_df.iloc[:,2:], norm='l1') 

        param_grid = {'C':np.logspace(-5, 8, 10)}   
        l1_svm = LinearSVC(penalty= 'l1', dual=False, max_iter = 5000) 
        grid = GridSearchCV(l1_svm, param_grid=param_grid, cv=5) 
        grid.fit(x_train_df_normalized,train_df.iloc[:,1])
        #score = grid.score(x_test_scaled,y_test_family)
        #sc.append(score)
        grid_pred_y_trian= grid.predict(x_train_df_normalized)
        #print(grid_pred_y_trian)
        #print(collections.Counter(grid_pred_y_trian))
        
        #print(i)
        #print("best parameter:",grid.best_params_)
        
        global train_accuracy
        train_accuracy = accuracy_score(train_df.iloc[:,1], grid_pred_y_trian) 
        #print("train_accuracy:",train_accuracy)
        global train_precision
        train_precision = precision_score(train_df.iloc[:,1], grid_pred_y_trian)
        #print("train_precision:",train_precision)
        global train_recall
        train_recall = recall_score(train_df.iloc[:,1],grid_pred_y_trian)
        #print("train_recall:",train_recall)
        global train_f1
        train_f1 = f1_score(train_df.iloc[:,1], grid_pred_y_trian)
        global train_auc
        train_auc = roc_auc_score(train_df.iloc[:,1],grid_pred_y_trian)   
        #print("train_auc:",train_auc)

    return train_accuracy,train_precision,train_recall,train_f1,train_auc
    
```
